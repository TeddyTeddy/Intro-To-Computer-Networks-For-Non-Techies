# Section 33 : Wired Network Topologies

There are four specific wired network topologies:

- Bus
- Ring
- Star
- Mesh



### Wired Network Topology : Bus topology

- Antiquated technology : Not used anymore
- All devices are connected to a single network cable
- Devices are connected to the bus via a T-connector
- Terminators are required at both ends of the cable to prevent signal bounce
- **Only one device can be active on the network at a time**
- The bus is the single point of failure

### Wired Network Topology : Ring Topology

- All devices are connected in a circular fashion
- Each device is connected to two other devices
- Data travels from node-to-node with each device handling data; either unidirectional or bidirectional
- Each device/node in the ring regenerates the signal, acting as a repeater
- Failure of a single node can take down the entire network
- **Fibre Distributed Data Interface** (FDDI) uses two counter rotating ring topologies for redundancy
- Each node will have 2 NICs (Network Interface Card) connecting to two other devices

### Wired Network Topology : Star Topology

The most popular wired network topology used is star topology

- All devices (e.g. printer, server or IoT) are connected to a switch
- Devices send data to the switch, which forwards it to the destination
- Can be applied to large or small networks
- The central switch is the single point of failure

### Wired Network Topology : Mesh Topology

- Each device is connected to every other device by seperate cable
- Higly redundant leading to fault-tolarent network
- Expensive to install
- Used in WANs and in Enterprises
- Partial mesh type (some level of mesh implemented, some connections missing)
- Full mesh type (as depicted in the picture above, where every device is connected with every other device)

## Lecture 34 : Wireless Network Topologies

Wireless networks utilize radio frequencies (RF) to communicate

There are three wireless network topologies, which are all currently in use:

- Ad hoc wireless network topology
- Infrastructure wireless network topology
- Mesh wireless network topology

### Ad hoc wireless network topology

Peer-to-peer (P2P) wireless network, where **no wireless access point (WAP)**
infrastructure exists

The P2P devices directly communicate with one another

PAN (Personal Area Networks) are a common example of Ad Hoc Wireless Networks

**Observation:** The above example network has **by arhitecture wise** a P2P architecture, **by size wise** is a PAN, **by topology wise** has ad-hoc wireless topology.

### Infrastructure wireless network topology

Wireless network that uses a wireless access point (WAP) as it's a central connecting device. WAP is connected to the wired network. Called also WLAN

Commonly used in homes & offices

### Mesh wireless network topology

Just like a wired mesh network topology, in wireless mesh network topology utilizes several wireless points (WAPs) to create a robust netword that is scalable (e.g. increase # of WAPs), self-healing and reliable. Common in businesses and larger homes.
Note that one of the WAPs is connected via a wire to a wired network. Then all the WAPs are connected to one another.

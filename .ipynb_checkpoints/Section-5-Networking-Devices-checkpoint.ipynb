## Lecture 37: Network Interface Cards (NICs)

NIC is the network adapter installed on your device.

It provides electrical, light or RF connections to the network media (e.g. cable, fiber or air respectively)

It can be either an expension card or build directly into the mother board

MAC addresses are burned into the NIC

You can have more than one NIC on a device

## Lecture 38 : Hubs

They are legacy equipment, no longer used, replaced by switches

OSI Layer 1 Device

Used to connect devices within a network

A hub is a multi-port repeater meaning that

- traffic goes in one port and is repeated (broadcasted out every other port)
- causes network collision errors
- Dumb network device due to multi port repeater nature

Due to the multi-port repeating mechanishm, it is much less efficient than a switch

## Lecture 39: Switches

They are a central connecting device connecting devices just like a hub

Intelligent Network Device at OSI Layer Layer 2 (Data Link Layer)

Memorizes the MAC Address if each device connected to it in **MAC address table**

Pays attention to Source & Destination MAC addresses during communications process

Breaks up collosion domains:

- Traffic goes in one port and is repeated out to only destination port
- Designed for High Bandwith
- Standard in todays network infrastructure

Switch is more secure than a hub because traffic goes in one port and is repeated out to only the destination port

When using a switch, we still have one large broadcast domain/area, in which we can broadcast to all devices. In case of retrieving a dynamic IP address, the new device can send a broadcast message to the broadcast domain saying "Hey, is there a DHCP server available, can this server assign me an IP address?"

## Lecture 40: Wireless Access Point

A wireless access point (WAP) is a bridge between the wired network & the wireless network. Just like a switch, it is a OSI Model Data Link Layer 2 device. Note that WAP is **NOT** a router. WAP can be connected to a switch

## Lecture 41: Wireless Range Extender

Extends the range of a wireless network by acting as a wireless repeater

Re-transmits radio signals from the wireless network it is associated with

Anything WAP sends gets repeated with the wireless range extender

## Lecture 42: Routers

Routers are used to connect different networks together

Routes traffic between networks **using IP addresses**

Routers use **routing protocols** to find the best way to get a packet of information from one network to another. Routing protocols are intelligent decision mechanisms. This makes the router intelligent.

**Routers break up broadcast domains**

OSI Layer 3 - Network Layer : TCP/IP Internet Layer Device

In OSI layer:
- Layer 3 = Router
- Layer 2 = Switch, **WAP**
- Layer 1 = Hub


Example of using a router: If a you are in one network and you want to check Gmail, which resides in a server within another network then the communication must go through router(s).

Example based on the picture above: At the very bottom of OSI Layer 1, we have the hub. It is a dumb device, if devices within the same network are connected via the hub. Then device A send data to the hub and the hub broadcasts the data to all other connected devices. There could be collosions during this process due to other devices sending data to the hub.

If we replace the hub with a switch, then we have a smart device making decisions to which port to send the data. This is possible because a switch has a MAC address table for every device connected to it.

If a device A sends data to a device X in another network; then A sends data to a switch, which then determines that the data is destined for another network. The switch will send the data to the router, which will use its routing protocols to get the data to the destination network. Then the data will be forwarded from router to the destination network's switch, and based on the MAC address of X, it forwards the data to X.

**Observation**: If you need to send data from one network to another, then we need both IP address & MAC address of the source & destination devices.

## Lecture 43 : Modems (Modulators & Demodulators)

Modems modulate one signal to another, such as analog to digital; 

modulation: An analog to digital conversion. A good example to analog/digital conversion is modulating a telephone analog signal into a digital signal that a router can understand.

demodulation: A digital to analog conversion. A good example is converting from digital voice signals in PC into analog signals for Internet / WAN telephone line.

Referring to the picture above, your Internet Service Provider (ISP) brings in the coaxial cable to your home. The cable brings in analog signal which is then modulated into a digital signal for your router (SOHO) to understand.

A SOHO device stands for Small Office Home Office device. SOHO device combines the router, DHCP server and a firewall. 

NOTE: Most devices enabling at home internet are SOHO devices, they are not WAPs.

## Lecture 46 : Firewalls

- Firewalls are the foundation of network security.
- Firewalls protect your network from malicious activity on the internet
- Firewalls preven unwanted network traffic on different networks from accessing your network by filtering data packets that go through them.
- They can be a standalone network device (network-based-firewall) or a software on a host (host-based-firewall).

Types of Firewalls

- Packet Filtering Firewalls: 1.st generation & the most basic. Basic filtering rules; we are going to allow packets from certain IP addresses & block packets from certain IP addresses (blacklisting). E.g. We are not allowing any traffic coming from HTTP port 80.
- Cirruit-Level Firewalls: 2nd Generation, monitors valid/invalid TCP sessions; a 3 way handshake is required to form a valid TCP sessions. Hackers use invalid sessions without a 3 way handshake.
- Application Layer 7 Firewalls (Next Generation Fire Wall (NGFW)): 3rd generation

## Lecture 47: DHCP Server

Most phone systems run over IP networks via dedicated protocols, such as the **Session Initiation Protocol (SIP)**

VoIP endpoint devices are hardware devices (phones) or software such as Cisco Jabber or Skype that allow you to make phone calls.

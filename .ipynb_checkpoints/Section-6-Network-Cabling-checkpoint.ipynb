## Lecture 52: Ethernet explained

Ethernet is a network protocol controlling how data is transmitted over a LAN. Ethernet as two aspects. 

It has a physical component: It supports networks built with coaxial, twisted pair and fiber optic cabling . It supports 10 Mbps speeds upto gigabit speeds depending on the cabling.

It has a logical component: Ethernet uses CSMA/CD (wired network & switched network) & CSMA/CA (wireless network) access methodology.

Referring to the physical component, the cabling, ethernet uses N-Base-T naming convention, where:

N: Signaling Rate
Base: a Signalling Type (Baseband communication)
X: Type of the cable

Example:
- 10Base-T: 10 Mbps Baseband Twisted Pair

## Lecture 54 : Twisted Pair Copper Cabling



4 twisted pairs of wires with RJ-45 connector

Why are they broken down into pairs?

- One cable has + signal & the other has - signal

Why are the pairs twisted?

- They reduce crosstalk & the noise from the environment

Downsides of Twister Pair Copper Cabling:

- Security concern is is signal emanations
- 100 Meters Maximum Therotical Distance


## Shielded vs Unshielded Twisted Pair Copper Cabling

Unshielded Twisted Pair (UTP): More succeptible to electromagnetic interference

Shielded Twisted Pair (STP): Less susceptible to EMI & Crosstalk (if each pair is shielded)

EMI (Electromagnetic Interference) : The disruption of the operation on an electronic device when it is in the vicinity of an electromagnetic field caused by another electronic device.

## Lecture 55 : Ethernet Physical Component : 568A and 568B Wiring Standards

If you are cabling your building, you want to pick one ethernet wiring standard for your building. If you pick both wiring standards for your building, the **pinouts** would be different for different parts of your building. Then, especially with older devices, you may run into trouble.

## Straight-Through vs. Crossover Cables

When cabling an area, we need to take into account the devices
we are connecting via a cable. Are they the same type or different types? 

If both devices are the same type (e.g. computer to computer), they would use the same pin numbers
for Tx (Transmitting) and Rx (Receiving). Then from device A to B, we need to cross over the Tx pair into Rx receiving end of device B.
That's why, we need to use **crossover cable** to connect the two.
Note that in crossover cable case, we use 568A on one side and 568B on the other side.

If both devices are of different type / unlike devices (e.g. switch to router), we connect the pinouts straight; the pinout is the same standard on each side. 



In Gigabit Ethernet, we use each pair for **both** sending (Tx) and receiving (Rx).

## The Plenum

It is the open space above the ceiling or below a raised floor.

All network cabling placed in the plenum should be "plenum-rated", because non-plenum-rated cabling are a fire hazard

## Lecture 58: Fiber Optic Cabling

Glass or plastic fiber that carriers light (photons are used instead of electrons)

Properties: 

- High bandwidth
- Long Distance
- Immune to EMI
- Does not emanate signals

Two types of fiber cabling:

**Multimode fiber (MMF):** Shorter distances up to 2 kms (LAN)

**Single Mode Fiber (SMF)**: Up to 200 kms @400 Gbps, WAN

Compared to twisted pair copper wiring, it is more expensive

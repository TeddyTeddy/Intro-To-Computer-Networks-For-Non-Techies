## Data Encapsulation & Decapsulation in the TCP/IP model (Lecture 14 of Bogdan)

When a frame is received in bits and bytes in the recipient, it goes to Data Link layer.
Data Link layer reads only the frame header & the frame trailer; Data Link layer does not check packet header, segment header or data. Via frame header and frame trailer, Data Link Layer decides that the frame has arrived without error or not. If the frame has arrived correctly, then data link layer strips the frame header and frame trailer off from the frame and forwards the packet to Network Layer.
Once network layer decides that the packet is for the recipient device, it strips of the packet header from the packet and forwards the segment to transport layer. Transport layer reads the segment header and it assembles the segments based on their segment number to form the data and forwards the data to the application layer.

## Lecture 86 : Network Access Methods

2 types of network access methods in Network Interface Layer:

**CSMA**
- Carrier Sense : checks network for communication
- Multiple Access: Multiple devices using the network
- Collosion Detection: Wired network
- Collosion Avoidance: Wireless network
- Multiple devices can be using the network at once!

We will talk about collosion detection in switching & switches course.

**Token Ring**
- The logical token is passed between devices on the network
- Only devices with the token can send data
- Token prevents network collisions
- Single Access: only a single device can access the network

## Lecture 87 : Address Resolution Protocol (ARP)

**Wrong**: Resides in Internet Layer in TCP/IP model

**Correct** The Address Resolution Protocol is a request-response protocol whose messages are encapsulated by a link layer protocol. It is communicated within the boundaries of a single network, never routed across internetworking nodes. This property places ARP into the link layer/network interface layer of the Internet protocol suite [2]

Resolves IP Address to MAC addresses (note the plural in MAC adressES)

Finds the hardware address of a host from a known IP address
and vice versa (RARP)

ARP Command : arp -a; the -a option tells the system to display the arp cache tables for all NIC interfaces on the computer.

**How ARP Works?**

Imagine computer 1 on the left is new to a local network. Note that every
device in the network has an **ARP table** containing IP address mapped to MAC addresses. Imagine that computer 1 has the IP address of computer 2 but it does not have the MAC address. So it **broadcasts an ARP request** accross all network devices. Only device 2 on the right having the right IP address responds back with **ARP response** containing the requested MAC address. Computer 1 saves the MAC address in its **ARP Table** and uses it for future application layer requests.

**ARP table** is stored in OS. If source device does have the IP address and the MAC address in the ARP table, then it does not send out an ARP request. Otherwise, it sends out an ARP request. 

The command for seeing the ARP table is **arp -a** on all operating systems

**Q&A for ARP Protocol**

1. When the switch needs to learn the destination mac address of a computer, does it also send an arp request to all the computers on the network asking for their mac address? Does a router also send an arp request when looking for a computer's ip address?

There is ARP and reverse ARP. You use ARP when  you know an IP address and need to find out what the associated MAC address is. Reverse ARP does the opposite. An ARP is essential a broadcast message, so yes it gets sent to all computers on the LAN. Routers don't use ARP, they use various routing protocols (we don't cover routing protocols in this course because they're well beyond the scope of a beginner course).

2. If a computer knows the mac address but not the ip address of a computer on another network, then will it send an arp request to every computer on every network or are arp requests only for LAN's?

It would use reverse ARP.

3. Are arp requests electrical signals that are sent from one computer to another?

Networks can use eletrical signals (copper cabling), light bursts (fiber optic cabling), or signals in the air (WiFi). So the answer is yes, if it is using copper cabling. If it uses fiber optic or WiFi, no.

4. When looking at the arp table in the command prompt, it shows the type as either static or dynamic but what does static and dynamic mean?

Refers to DHCP. If the IP address associated with the MAC address is assigned by DHCP, it is considered dynamic. If it is statically assigned by someone, is shown as static.

5. In the arp table in the command prompt, are all the ip and mac addresses shown there from computers that our computer has communicated with?

Yes

6. Can multiple device connected to a same IP?

No, every device on a network must have a unique IP address.

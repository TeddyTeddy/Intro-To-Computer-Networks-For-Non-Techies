## Bandwidth (Lecture 23 of Bogdan)

Bandwith is max. amount of bits that could be transmitted over the link in specific amount of time.

1Kbps = 1000 bits/sec

1MBits/sec = 1 000 000 bits/sec

1 GBits/sec = 1 000 000 000 bits/sec

Note that bandwidth is **the characteristic of physical media**

## Throughput (Lecture 24 of Bogdan)

Actual amount of data that could be transmitted over a specified time is called throughput.

Like bandwith, throughput is measured in bits/sec as well.

Imagine a 100 MBit/sec bandwidth connection on a video call.

We want to calculate what the throughput is.

Remember that we have Encapsulation in TCP/IP Model:

If we want to calculate **the throughput at the application layer:**

Imagine that data has a size of 10 KB and the frame has a size of 12 KB.
Then the throughput can be calculated as: (10 / 12)*100 MBit/s = 83.3 MBit/sec

In short, thoughput tells how many bits/sec of data can be transferred over the network

If we want to calculate **the throughput at the network layer:**
then you need to:

(packet size / frame size)*bandwidth 

Note that network layer throughput will be higher than the application layer throughput. Similarly throughput for the application layer is less than the transport layer, which is less than the network layer, which is less than the data link layer.


## Latency (Lecture 26 of Bogdan)

The time it takes for data frames to arrive from one device to another (one-way-delay)

Reasons for latency:

- **Queueing delay** : If there are multiple conversations ongoing (e.g chrome, video streaming) while sending an email, NIC is busy sending & receiving the packets.In such case the email frames are put into a queue along with video streaming frames in NIC. The time it takes for email frames to be sent is called queueing delay. If you have low bandwith connection, then you might have high queueing delay.
- **Processing delay**: The time it takes at each layer of each device on the connection path.

- **Transmission delay**: How much time it takes for each frame to be transmitted over wired or wireless connection. This delay depends on the bandwith of the media. The higher the bandwith the less the transmission delay is or vice versa.

- **Propagation delay**: how much time time it takes for a single bit from one end to another.

At which layer of TCP/IP do we measure latency?

Application layer.

## Round-trip Time (RTT) PING (Lecture 27 of Bogdan)

Time required to send data from one device to another and back.

RTT lower than 300ms is almost unnoticable in audio & video calls.

## Jitter (Lecture 29 of Bogdan)

Jitter happens when different packets have different one way delays. As a result, packets will arrive in a different order than they were sent.

Imagine a video transmission for the example above. Due to packets having different one-way delays, the packets arrive at the receiver in different order. It is responsibility of TCP protocol; it is capable of reordering the packets at the receiver. 

In summary, **jitter is due to variable latency in different packets**

## Packet Loss Rate (Lecture 30 of Bogdan)

Due to congestion, link failures or processing failures packets can be lost on the way. It is the responsibility of TCP to re-send the lost packets.

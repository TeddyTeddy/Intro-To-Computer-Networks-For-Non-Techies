## Disadvantages of switched network (Lecture 47 of Bogdan)

Lets assume that we have a large switched network where many switches are interconnected!

Disadvantages of a switched network:

- (1) When the destination MAC address is **not** in the MAC address table, the switch sends broadcast frames with broadcast MAC address to all ports, except the source port. This means that the frame will be broadcasted within all ports except the source ports in all switches. This means the broadcast frame floods the entire switched network (a.k.a. broadcast storm)

- (2) Large MAC address tables: If there are many switches in the network, then each switch will have very large MAC address tables due to being connected to other switches. Also, each time a new device is connected to a new port, there will be MAC address table updates accross all the switches. In addition, if there is an unknown MAC address, then disadvantage (1) will occur, floating the network.

If you want to build CAN/MAN/WAN then switched network solution becomes unfeasible.

## What is IP address (Lecture 48 of Bogdan)

IP address is a logical address of the device in the network layer

IP address may change but MAC address usually remains unchanged

IP address is usually dynamically assigned by DHCP servers, IP addressess can be assigned statically as well.

MAC addresses are represented using hexadecimal characters, where as IP addresses are represented using decimal characters.

## IPv4 vs. IPv6 addresses (Lecture 49 of Bogdan)

There are IPv4 protocol and IPv6 protocol defining the internetwork communications in internet layer in TCP/IP model.

IPv4 address is in decimal format
IPv6 address is in hexadecimal format (128 bits - 8 sections - 4 hex character in each section)

**IPv4 Address format**

Max value of each octet = 1111 1111 in binary = ((2 power 8) - 1) in decimal = 256 - 1 = 255
Min value of each octed = 0

So, each octet can have value from 0 to 255.

Max # of IP addresses = ((2 power 32) - 1) = 4 294 967 296

How to manage internetwork connectivity with such limited # of IP addresses?

There are 2 different techniques to manage connectivity with a limited # of IP addresses:

- Private and Public IP addresses
- NAT (Network Address Translation)

## Types of IPv4 addresses (Lecture 52 of Bogdan)

**Unicast IP addresses** are assigned to specific devices in the network. The assigment can be done dynamically via DHCP or statically by manually assigning an IP address.

**Multicast IP addresses** are when a source device wants to send packets to a specific group of devices. Note that Multicast IP address starts with **224**.

**Broadcast IP addresses** are assigned to 255.255.255.255. Broadcast IP addresses are used when a source wants to send packets to all devices in the same switched network. In the same switched network broadcast IP packets do **not** travel all over the world.

## What is router? (Lecture 53 of Bogdan)

**Important:** In the above picture, layer 3 of TCP/IP model should be named as Internet Layer, not network layer.

Router is network device responsible for routing IP packets on network layer of TCP/IP layer. That means that router works on physical, data link and network layer. Note that router has ports where each port has a MAC address and a IP address. 

Note that routing IP packets is not the only responsibility of a router. Another responsibility of the router is to discover remote networks. There are **network discovery protocols** enabling routers to discover networks of each other. They build **routing tables** and forward IP packets in layer 3 of TCP/IP model (internet layer).

## Local Area Network (LAN) vs. Wide Area Network (WAN) (Lecture 54 of Bogdan)

LAN is a set of devices and interconnections under common administration located in the same geographical area (e.g. room, floor etc)

WAN covers large geographical areas and consists of multiple LANs.

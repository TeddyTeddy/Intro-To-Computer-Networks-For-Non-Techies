## IP Address Classes (Lecture 56 of Bogdan)

**Important:** Refer to First Octet column in the table; when analzing IPv4 address classes, we number the bits from left to right. Let's analyze each IPv4 address class:

- For class A: 0 on the left is the first bit the rest goes from there
- For class B: 10 on the left are the first two bits; the first bit is 1 and second bit is 0 and so on..
- For class C: 110 on the left are the first three bits; the first bit is 1, the second bit is 1 and the third bit is 0 and so on..
- For class D: 1110 on the left are the first four bits; the first bit is 1, the second bit is 1 and the third bit is 1 and the fourth bit is 0 and so on..
- For class E: 1111 on the left are the first four bits; the first bit is 1, the second bit is 1 and so on..

Tip: If you want to check what IP address class your IP address is, check the first octet of the 
in binary form:

> ipconfig /all

IPv4 Address. . . . . . . . . . . : 192.168.43.117(Preferred)

First Octet 192 in binary form 11000000, which is 110xxxxx
So; the IPv4 address is of class C.

** Note that the second, the third and the fourth octets can have any values.



Note that class D is multicast IP address that can only be used 
as destination IP addresses in a frame. Multicast IP addresses are used for example for video and audio streams.

class E type IP address is reserved for future use.

In short, end users' devices' IP addresses are of type A, B or C.


Note also that between class A and class B type IPv4 addresses, there is a gap;
127.0.0.0-127.255.255.255. In the next lecture, we will explain Localhost/loopback IP address.

## Localhost IP address: 127.0.0.1 (Lecture 57 of Bogdan)

127.0.0.1 is reserved as localhost IP address to test if TCP/IP stack is working 
for local device. Also using localhost IP address, you can access services running on local host.

## Private vs Public IP address

Private IP addresses are the ones that are not routable in global internet; that means that routers of ISPs will drop any packets whose source/destination address is a private IP address (*).

Referring to the picture, the devices belonging to the same network (i.e. their addresses start with 192.168.1.x) can communicate with each other directly. But when any of the devices in the private network wants to communicate with internet directly with its private address, then the access will be blocked (due to (*)). But if the router, which has the private network, has a public IP address, then the router can access resources in internet. 

We can copy the private network with its private IP addresses as many times as we want. Below given an example;

## Private IP address ranges (Lecture 61 Bogdan)

/8 is the default network mask for class A networks

/16 is the default network mask for class B networks

/24 is the default network mask for class C networks

If you want to create a private LAN, you can chose any of the ranges from the above private
IP address ranges such as;

- 10.0.0.8/8     for a private class A network
- 172.16.0.0/16  for a private class B network
- 192.168.1.0/24 for a private class C network

note that the above example address blocks can be used any number of times in internet

## Public IPv4 Address Ranges (Lecture 62)

Public IP addresses can be calculated as:

- Any class A address that is not private
- Any class B address that is not private
- Any class C address that is not private



You cannot assign a public IP address to your device. You need to get permission from RIPE organization to do that.

Having a public IP address costs money.

## Static IP addresses (Lecture 63 of Bogdan)

Static IP addresses are assigned permanently to servers running a public services such as DNS or SMTP service.

DNS is used to map your hostname to the static IP address permanently

An example public static IP address is 8.8.8.8 -> Googles DNS servers

Requires additional security; once attackers know your static IP address.
That's why one needs to setup firewalls opens only the necessary ports for connections.

End user devices such as phones, laptops etc have do NOT have static IP addresses but
they do have Dynamic IP addresses.

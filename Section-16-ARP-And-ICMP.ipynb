when two hosts want to communicate.

ICMP can perform connectivity check and other tasks between two hosts.

After that we will see examples of how two different hosts can communicate
between each other when located in the same network or in different networks.

## How are packets sent to remote hosts? (Lecture 96 of Bogdan)

As stated in the picture, a source host with IP 172.16.1.5 wants to send an IP packet to 8.8.8.8 (Google's DNS server). Note that the DNS server is somewhere behind the router in internet. 

Q) Will the source host with IP 172.16.1.5 send ARP Request to broadcast MAC address?
A) No, because the source & destination hosts are not in the same sub-network;

172.16.1.5  when masked with /24 -> 172.16.1.0  (source host's network address)
8.8.8.8     when masked with /24 -> 8.8.8.0     (destination host's network address)

In this case, the source host decides not to send an ARP request. 
Instead, the source host will forward the packet to its default gateway:
in data link layer in destination mac address field (in the frame), it will indicate
destination mac address as the default gateway's mac address.

Default gateway will notice that the packet is targeted to itself 
(i.e. destination mac address is its own mac address), performs decapsulation, looks into routing table finds the route(s) and sends the packet to the route etc.

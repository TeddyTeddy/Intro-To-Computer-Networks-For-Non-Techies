#  What is a computer network?

**Lecture 11 of [1] & Lecture 3 of [Bogdan]:**

"A network is a set of devices sharing resources with one another via a set of connections between them. When we mean internet, nowadays we mean different devices connected to one another; Internet of Things (IoT) in short"

A network is composed of two main aspects:

a. Physical connection (wired, wireless)

b. Logical connection (data transporting accross the physical media via electrical signals through protocols)

In a network, protocols are used to send a resource (e.g. a txt file) from one computer to another.

# Basic Networking Rules

Lecture 5 of [Bogdan]:

There are different models (e.g. TCP/IP or OSI Model) that describe how devices talk to one another.

 Lecture 11 of [1]:

1. The computers in a network must use the same procedures for sending & receiving data.

   Observation: The procedures for sending and receiving data are called network protocols.
   
   In other words, the computers in the network must use the same communication protocols.

2. Data must be delivered uncorrupted. If it is corrupted, it is usually useless, but there are exceptions
   such as in VoIP.

3. Computers in a network must be capable of determining the origin & the destination 
   address of the data sent. Those are the IP address & the MAC address.

# Lecture 12 : Types of Networks By Size

### 1. Personal Area Network (PAN)

Ultra small networks used for personal used to share data from one device to another.

Examples: Smartphone to Laptop, Smart Watch to Smart Phone, hands free device & smartphone

### 2. Local Area Network (LAN)

A computer network within a small area such as a room, floor or building.

Examples: Home network, office network

**Figure 1** - A LAN network

Size can vary for a LAN depending on the number of devices connected. It can be a single room or a collection of
rooms or whole floor or whole building(s)

## 3. Wireless Local Area Network (WLAN)

A LAN that's fully dependent on wireless connectivity.

## 4. Campus Area Network (CAN)

A network of multiple interconnected LANs in a limited area, such as a 
business park, goverment agency or university campus

## 5. Metropolitan Area Network (MAN)

A computer network that interconnects users with computer resources in a city.
Larger than campus area network but smaller than wide area network (WAN)

## 6. Wide Area Network (WAN)

A network that extends over a large geographical area spanning over multiple cities & countries.

Examples: Google offices in different states, the internet in US

Summary of networks by size: WAN > MAN > CAN > WLAN > LAN > PAN

# Lecture 13 : NETWORK ARCHITECTURE

Two types of network architectures

## (1) Peer to Peer Network Architecture

**Peer:** an equal, a match, a like

All devices on the network are peers (equal in status); **In a P2P network, there are no dedicated servers** so there is no centralized control over shared resources.

Any device can share its resources as its pleases. Any device can act either as a client or server.

## (2) Client Server Network Architecture

The network is composed of clients & servers. These roles are fixed. 
Clients receive resources and servers provide resources.

Servers provide centralized control over resources (e.g. printers, files etc).

Servers centralizes user accounts, security and access controls.

In the above picture, if the client provides the correct credentials to the authentication server, 
it will get access to the files (i.e. resources) on the file server

**IMPORTANT** When we look at the **network architecture types**, we look not only at equipment types but also
**how the devices access resources on the network**. The decentralized method is called P2P Network Architecture. The centralized method is called client & server network architecture.

**Observation:** When trying to analyze a network, first question to ask is what is the network architecture.

Compared to client & server network architecture, a P2P network architecture does **not** have centralized access control, user accounts and security.

# Lecture 18: Introduction to Computer Networking Protocols

## Protocols: 

Protocols are 

    - the rules that govern data exchange and enable network communication.
    - the procedures for sending and receiving data


Devices communicate with one another through the use of protocols.

Ex: When you pick up the phone, you check if there is a dial tone and then you dial the number. When the other side picks up the phone, you say 'Alo' as the first word and so on.

Ex: When you drive a car, you obey the traffic rules

**Observation:** A protocol can take care of a sub-task enabling network communication. 

### Protocols are devided into 2 parts:

1. **Physical protocols** describing the medium (the wiring), the connections (RJ-45 port) and the signal (voltage & current level on the wire)

2. **Logical protocols**: the software controlling how and when data is sent and received, supporting physical protocols

Computer networks depend on a chain of physical & logical protocols to work properly.

Protocols handles a specific task in networks. For example:

    - File transfers : FTP
    - Email: POP3, SMTP, IMAP
    - Web Communication / Browsing: HTTP


# Lecture 19 : Introduction to the OSI model

### What is OSI model?

OSI stands for Open Systems Interconnection Reference model.
It is a **conceptual** framework showing us how data moves throughout
a network. It's purpose is to show how networks operate.
It is not implemented in the real world, but TCP/IP is.



### The OSI Model Stack

The OSI Model has 7 layers. 

**Lower layers** (Media layers) provide network specific functions such as addressing, routing and flow control

**Upper layers** (Host layers) are managed my the host computer and perform application specific functions such 

as data formatting, encription and connection management.

# Lecture 21 : Media Access Control (MAC) Address

Media Access Control (MAC)

 - Physical address of the network adapter / network interface card (NIC)
 - OSI Layer 2 (Data Link) Layer Address
 - TCP/IP (Network Interface) Layer Address

Six bythes (48 bits), usually represented in Hex format

First three bytes (24 bytes) are assigned by the IEEE to the manufacturer as Organizationally Unique Identifier (OUI)

Last three bytes (24 bits) are unique

Example:
00:21:70:6f:06:f2
                    
2 power 24 = 16.7 million unique OUIs and 16.7 million unique devices for each OUI

## Lecture 22 : Introduction to IP addresses

An IP address is a **logical address** used in order to **uniquely identify**
a device on an IP network

Unlike a hard coded MAC address, we have to actually assign it on OS in the internet layer

**IP address is associated with routing**. It is used in OSI Layer 3 : Network Layer
and TCP/IP Layer 2 : Internet Layer

There are two versions: IPv4 and IPv6 addresses


## Lecture 23 : IP address vs. MAC address

| IP addresses | MAC addresses |
| :- | :- |
| Network OSI Level 3 addresses | Data Link Layer OSI Layer 2 addresses |
| Logical addresses | Physical addresses |
| Assigned in Operating System | Physically burned on NIC |
| Allows network to network communication via routers | Allows internetwork communication via hubs, switches and routers |
| WAN communication | Local LAN communication |

## Lecture 24 : Half vs Full Duplex Communication

Network Communication will occur in either full or half duplex mode:

- Half duplex mode: Can send or receive data, but not at the same time
- Full duplex mode: Can send and receive data at the same time

## Lecture 72 : Why do we plan a network?

Many businesses utilize an ad-hoc network design with a break-fix mentality

Ad Hoc Networks with a break-fix mentality can lean to issues:

- Lack of documentation
- Troubleshooting issues
- Scalability issues
- Reliability issues
- Security issues
- Data loss issues
- Costly repeairs

Good network planning reduces business downtime

## Lecture 73 : The overall process

At planning process, we look at the business needs & the technical needs which yield our initial requirements to design the network. Depending on the project constraints (i.e. time, skills, quality) we design the network.
The design network gets implemented and put into production, where it gets maintained over time. Then over time, over long term, then we need to tune 
it by upgrading its hardware & software & configuration. The process is cyclical.

## Lecture 82 : Network Planning Exercise

Network Planning Process Student Exercise

Assume you're an IT consultant and you've been contacted by a potential client
to design and build out their network infrastructure. Before you visit the client, 
formulate four general network design planning questions you would ask a potential
client based on the vague scenario below.

“A small civil engineering firm is experiencing a lot of recent business growth and
has identified a need to expand its business into a second office location across town
that they just signed the lease for yesterday. They don’t have any in-house IT expertise,
so they’ve contacted you, and are requesting a Request for Proposal (RFP) to design
and build the network at their new office location."

I try to ask the client business related questions. Most clients will have hard time
understanding the technical aspects of the network. One feasible way to get the necesssary technical
information (e.g. data security via encription) is to ask them the desired level of security
(e.g. none, bare minimum, medium, high, strongest possible) and translate the answer into
techical requirement (e.g. encription method A, B or C).

Business Analysis Questions:

0. What is the business model?
   https://www.emergn.com/wp-content/uploads/2017/01/business_model_canvas.png
1. What are the key business area(s)?
2. For each business areas, what are the key processes?
3. For each key business process, 
	- Who are the key people & what are their roles?
	- What are the applications / tools / systems?
	- What are the regulations per key business process?
	- What sort of data is created/used/transferred within the key business process?
4. Are there any regulations that are valid accross all the key business processes?
5. What are the roles in the second office location? 
   How many people per role (maximum) will be using the network?
6. What is the budget? 
7. How long is the project supposed to last?
8. Which business processes are already automated?
9. Which business processes can be newly automated?
10. Should services (e.g. mailing, printing) should the interconnected between seperate office networks?
11. Tell me about the new office location. How big is it? 
    How far is it from the current office? Is it one big office space or broken up into smaller rooms? 
	Is it on one floor or many floors? Are there suspended floors and ceilings?
12. What sort of office tools do they need eg internal email? 
    Server storage? Access to video conferencing? 
	Do you want the telephony to work over the network?
13. Where do you want to store your data? 
    For how long will you store your data?
	Do you use a version control system (e.g. Git) to store different versions of your data?

Technical Analysis Questions
1. What is the expected availability rate? 
2. Is data encription needed within the network? 
   If so, what should be the desired level of security?
3. What is the minimum required speed? 
4. What type of cabling is preferred? UTP, STP, Fiber optic, Wireless
5. Is printing services required?
   a. How many printers?
   b. How many printers will be shared per floor? What type of printers will be shared?
   c. Which cubicles will have their own printers?
6. Do you need your data to be served from a dedicated centralized server and be consumed by several clients? (e.g. client server architecture)
   Or should each device in the network decide what to share with whom and when? (peer to peer network)

## Introduction

After we had a high level overview on the OSI model and the TCP/IP model, now it is time time to review Ethernet, which is the most popular protocol that runs in the data link layer & the physical layer. Using ethernet, you can transfer data through wired media or wireless media.

## What is Ethernet? (Lecture 16 of Bogdan)

Ethernet is a set of standards & protocols for data link and physical layers.

Ethernet is not a single protocol; it is a set of different protocols in data link & physical layers.

Ethernet also encompasses IEEE 802.3 (for wired media like copper & fiber optic) and IEEE 802.11 (for wireless communication) standards.

## Copper, Fiber and Wireless Media in Ethernet (Lecture 17 of Bogdan)

Copper cables:

- cheap
- Standardized RJ-45 connectors
- Different cable types
- Different supported speeds up from 100Mbps - 10GBps
- Max Length is 100 m

Fiber Cables:
- Higher transmission rates than copper
- Max cable distance is 100 km
- Less succeptible to external physical factors
- Expensive especially single mode cables and transreceivers

Disadvantages of wireless media:

1. Interference - When multiple devices transmit signal on the same channel simultaneously.
2. Wireless distances are limited


## Network Interface Card (NIC) : Lecture 18 of Bogdan

We have just discussed that Ethernet supports different media types on
physical layer of the TCP/IP model such as copper cables, fiber cables or wireless.

If a device wants to communicate through a media, it must have a specific network interface controller (NIC) for the media in question.

Every NIC has built-in MAC address. This address is called MAC address.

## Media Access Control (MAC) address (Lecture 19 of Bogdan)

Every NIC has a hard coded (fixed) MAC address

Six bytes (48 bits), usually represented in Hex format

First three bytes (24 bytes) are assigned by the IEEE to the manufacturer as Organizationally Unique Identifier (OUI)

Last three bytes (24 bits) are unique

Example:
00:21:70:6f:06:f2
                    
2 power 24 = 16.7 million unique OUIs and 16.7 million unique devices for each OUI
